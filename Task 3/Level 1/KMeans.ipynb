```python
import cv2 as cv
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
img = cv.imread('/content/drive/Shareddrives/AI/sample.jpeg')
img=cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
img_data=(img/255.)
img_data=img_data.reshape(-1,3)
```

```python
kmeans =  KMeans(2).fit(img_data)
k_colors = kmeans.cluster_centers_[kmeans.predict(img_data)]
```

```python
img.shape
```

```python
import numpy as np

k_img = np.reshape(k_colors,(img.shape))
final=np.ones((600,750,3))
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(final,'K = 1',(320,50), font, 1, (0,0,0), 2, cv.LINE_AA)
final[100:567,25:725,:]=k_img
plt.imshow(final)
```

```python
import imageio
import numpy as np
images=[]

arr=[1,2,3,4,5,6,8,10,15,25,45]
for k in arr:
    kmeans =  KMeans(k).fit(img_data)
    k_colors = kmeans.cluster_centers_[kmeans.predict(img_data)]
    
    k_img = np.reshape(k_colors,(img.shape))
    final=np.ones((600,750,3))
    font = cv.FONT_HERSHEY_SIMPLEX
    cv.putText(final,'K = '+str(k),(320,50), font, 1, (0,0,0), 2, cv.LINE_AA)
    final[100:567,25:725,:]=k_img
    plt.imshow(final)
    images.append(final)
    images.append(final)
    images.append(final)
    images.append(final)
    images.append(final)
```

```python
imageio.mimsave('/content/drive/Shareddrives/AI/result.gif', images)
```
